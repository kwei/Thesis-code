```python
model = 'VAE'
```

```python
from functools import reduce
from operator import add
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
from matplotlib.pyplot import MultipleLocator

fig, ax1 = plt.subplots(1,1, figsize = (8,5))

labels = []

binSize = 50000

patterns = ["", "/" , "x", "O", ".", "*" ]
plotMarkers = ['', 'o','x','^']
colors = ["","#FF000099","#00000099","#0000FF99"]

matplotBars = []
barWidth = 0.15
padding = 0
picks = [1, 2, 3]
divs = [50]

socp_after_rmse_index_90_list = []
for pick in picks:
  for div in divs:
    f = open(rootpath+'/datas/localizationResult/1213_after_localization_RMSE{}_{}_{}.pkl'.format(div, pick, model), "rb")
    afterRMSE = pickle.load(f)
    f.close()
    After_sumRMSE = np.array(reduce(add, afterRMSE))
    After_sumRMSE = After_sumRMSE[After_sumRMSE < np.max(After_sumRMSE)*0.9]
    socp_after_rmse_index_90_list.append(np.mean(After_sumRMSE))
x = np.arange(0,len(socp_after_rmse_index_90_list), 1)
_bar = ax1.bar(x-barWidth-padding,socp_after_rmse_index_90_list, label=r'SOCP', zorder=3, align='center', width=barWidth, edgecolor='black', hatch=patterns[1])


sdp_after_rmse_index_90_list = []
for pick in picks:
  for div in divs:
    f = open(rootpath+'/datas/localizationResultSDP/0808_3_after_localization_RMSE{}_{}_{}.pkl'.format(div, pick, model), "rb")
    afterRMSE = pickle.load(f)
    f.close()
    After_sumRMSE = np.array(reduce(add, afterRMSE))
    After_sumRMSE = After_sumRMSE[After_sumRMSE < np.max(After_sumRMSE)*0.9]
    sdp_after_rmse_index_90_list.append(np.mean(After_sumRMSE))
x = np.arange(0,len(sdp_after_rmse_index_90_list), 1)
_bar = ax1.bar(x,sdp_after_rmse_index_90_list, label=r'SDP', zorder=3, align='center', width=barWidth, edgecolor='black', hatch=patterns[2])


after_rmse_index_90_list = []
for pick in picks:
  for div in divs:
    f = open(rootpath+'/datas/localizationResult/0804_after_localization_RMSE{}_{}_{}.pkl'.format(div, pick, model), "rb")
    afterRMSE = pickle.load(f)
    f.close()
    After_sumRMSE = np.array(reduce(add, afterRMSE))
    After_sumRMSE = After_sumRMSE[After_sumRMSE < np.max(After_sumRMSE)*0.9]
    after_rmse_index_90_list.append(np.mean(After_sumRMSE))
x = np.arange(0,len(after_rmse_index_90_list), 1)
_bar = ax1.bar(x+barWidth+padding,after_rmse_index_90_list, label=r'SD/SOCP', zorder=3, align='center', width=barWidth, edgecolor='black', hatch=patterns[3])


ax1.grid(zorder=0)
font_size = 20

ax1.legend(loc="upper left", prop={'size': font_size}, ncol=1)
# ax1.set_xlabel('Anomaly', fontsize=font_size)
ax1.set_ylabel('RMSE [m]', fontsize=font_size)

ax1.set_ylim([0, 1200])
# ax1.set_xlim([0, 3])

# ax1.set_ylim([0,1200])
# ax1.set_yticks(np.arange(0, 1200, 300))
# ax1.set_yticklabels([str(x) for x in range(1200) if x%300 == 0])
# plt.yticks(np.arange(0, 1200, 300), fontsize=font_size)
# ax1.yaxis.set_ticks(np.arange(0, 1200, 300))
y_ticks = [x for x in range(1200)]
y_labels = []
for x in y_ticks:
  if x%300 == 0:
    y_labels.append(str(x))
  else:
    y_labels.append("")

plt.yticks(np.arange(0, 1300, 100), [str(x) for x in np.arange(0, 1300, 100)], fontsize=font_size)
plt.xticks([0, 1, 2], ["$M={}$".format(pick) for pick in picks], fontsize=font_size)

# ax1.xaxis.set_major_locator(MultipleLocator(1))
# ax1.yaxis.set_major_locator(MultipleLocator(50))

plt.savefig('localization_error_comparison_SOCP_SDP_SDSOCP_RMSE.eps', bbox_inches='tight')
# files.download('localization_error_comparison_SOCP_SDP_SDSOCP_RMSE.eps')

# plt.savefig('localization_error_comparison_SOCP_SDP_SDSOCP_RMSE.png', bbox_inches='tight')
# files.download('localization_error_comparison_SOCP_SDP_SDSOCP_RMSE.png')
```
